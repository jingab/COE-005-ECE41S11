```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading of data
df = pd.read_csv('data.csv')
df.info()
```

```python
#Storing of the parameters/features
P = df.iloc[:,2:32] #Columns 2-32 will be the parameters/features
print(P)
```

```python
#Storing of the target variable
D = df["diagnosis"] #column for diagnosis will be the target variable
print(D)
```

```python
# Splitting of the dataset into training set and test set
from sklearn.model_selection import train_test_split # Import train_test_split function
P_train, P_test, D_train, D_test = train_test_split(P, D, test_size=0.2, random_state=110) # 80% training and 20% test
```

```python
#Defining the model to be used
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(P_train, D_train)

#Predict the response for test dataset
D_pred = clf.predict(P_test)
print(D_pred)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(D_test, D_pred))
```

```python
# Model Precision: what percentage of positive tuples are labeled as such?
print("Precision:",metrics.precision_score(D_test, D_pred,pos_label='positive',average='weighted'))

# Model Recall: what percentage of positive tuples are labelled as such?
print("Recall:",metrics.recall_score(D_test, D_pred,pos_label='positive',average='weighted'))
```

```python
#Testing of the model
testing = pd.read_csv("test.csv") #storing data of test csv to testing
testing.head()
```

```python
#Prediction of the testing data
testing_pred = clf.predict(testing)
print(testing_pred)
```
